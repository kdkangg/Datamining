```python
import pandas as pd
import numpy as np
```

```python
#read data
calories = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/calories.json')
distance = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/distance.json')
#l_act = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/lightly_active_minutes.json')
m_act = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/moderately_active_minutes.json')
#r_heart = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/resting_heart_rate.json')
sedentary = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/sedentary_minutes.json')
sleep_score = pd.read_csv('../osfstorage-archive/pmdata/p12/fitbit/sleep_score.csv')
sleep = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/sleep.json')
steps = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/steps.json')
hr_zone = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/time_in_heart_rate_zones.json')
v_act = pd.read_json('../osfstorage-archive/pmdata/p12/fitbit/very_active_minutes.json')
```

```python
#data frequency to date using sum
calories_processed = calories.resample('D', on = 'dateTime').value.sum() 
distance_processed = distance.resample('D', on = 'dateTime').value.sum()
steps_processed = steps.resample('D', on = 'dateTime').value.sum()
```

```python
#resting heart rate to simple df
#r_heart['value']=r_heart['value'].apply(lambda x: x['value'])
#r_heart_processed=r_heart
```

```python
#delete time in sleep score time stamp
sleep_score['timestamp']=sleep_score['timestamp'].apply(lambda x: x[:10])
#drop unnecessary column
sleep_score_processed = sleep_score.drop(columns='sleep_log_entry_id')
```

```python
#dictionary with heart rate zone minutes to df
hr_zone_value = pd.DataFrame([eval(str(a).replace('{\'valuesInZones\': ','').replace('}}','}')) for a in hr_zone['value']])
hr_zone_processed = pd.concat([hr_zone['dateTime'], hr_zone_value], axis=1)
```

```python
#unify index
#l_act_processed = l_act.set_index('dateTime')
m_act_processed = m_act.set_index('dateTime')
#r_heart_processed = r_heart_processed.set_index('dateTime')
sedentary_processed = sedentary.set_index('dateTime')
sleep_score_processed = sleep_score_processed.set_index('timestamp')
sleep_processed = sleep.set_index('dateOfSleep')
hr_zone_processed = hr_zone_processed.set_index('dateTime')
v_act_processed = v_act.set_index('dateTime')
#index to DateTimeIndex
sleep_score_processed.index = pd.to_datetime(sleep_score_processed.index)
```

```python
#get dummy values for mainsleep true(1)/false(0)
sleep_processed['mainSleep'] = pd.get_dummies(sleep_processed['mainSleep'])[True].fillna(0)
#drop unnecessary columns
sleep_processed = sleep_processed.drop(['logId','startTime','endTime','type','infoCode','levels'], axis=1)
#index to DateTimeIndex
sleep_processed.index = pd.to_datetime(sleep_processed.index)
```

```python
#read prediction target
wellness = pd.read_csv('../osfstorage-archive/pmdata/p12/pmsys/wellness.csv') 
wellness['effective_time_frame'] = wellness['effective_time_frame'].apply(lambda x: x[:10])
wellness = wellness.rename(columns={"effective_time_frame": "dateTime"})
readiness = wellness.loc[:,['dateTime','readiness']].set_index('dateTime')
readiness.index = pd.to_datetime(readiness.index)
```

```python
#drop duplicated records
readiness, calories_processed,distance_processed,m_act_processed,sedentary_processed,sleep_score_processed,sleep_processed,steps_processed,hr_zone_processed,v_act_processed = [x[~x.index.duplicated(keep='first')] for x in [readiness, calories_processed,distance_processed,m_act_processed,sedentary_processed,sleep_score_processed,sleep_processed,steps_processed,hr_zone_processed,v_act_processed]]
```

```python
#concat data
P12 = pd.concat([readiness, calories_processed,distance_processed,m_act_processed,sedentary_processed,sleep_score_processed,sleep_processed,steps_processed,hr_zone_processed,v_act_processed], axis=1)
```

```python
#rename columns
P12.columns=['readiness','calories','distance','m_act','sedentary','sleep_overall_score','sleep_composition_score',
             'sleep_revitalization_score', 'sleep_duration_score',
             'deep_sleep_in_minutes', 'sleep_resting_heart_rate','sleep_restlessness',
             'sleep_duration','sleep_minutesToFallAsleep','sleep_minutesAsleep','sleep_minutesAwake',
             'sleep_minutesAfterWakeup','sleep_timeInBed','sleep_efficiency',
             'main_sleep','steps',
             'hr_below_zone1', 'hr_zone1', 'hr_zone3','hr_zone2','v_act']
```
